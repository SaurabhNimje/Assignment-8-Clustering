#Kmeans

```python
#Import libraies
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
from google.colab import files
uploaded=files.upload() #Upload file
```

```python
data1=pd.read_csv("crime_data.csv") #Read file
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(data.iloc[:,1:])
```

##ELBOW Graph

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#From ELBOW Graph we take K=3

```python
# selecting 3 clusters from above scree plot
model=KMeans(n_clusters=3)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
data1['new_clusters']=x
data1
```

```python
data1.groupby('new_clusters').agg(['mean']).reset_index()
```

```python
data1[data1['new_clusters']==0] #0th Cluster
```

```python
data1[data1['new_clusters']==3] #In 3rd Cluster no value are occure bcz we are only 3 clusters
```
